```python
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
data = pd.read_csv('dataset_diabetes/diabetic_data.csv', na_values='?')
data
```

```python
# Convert readmitted Variable to 0 and 1
data.readmitted[data.readmitted == 'NO'] = 0
data.readmitted[data.readmitted == '>30'] = 0
data.readmitted[data.readmitted == '<30'] = 1

# Unique values and counts for readmitted
print(data.readmitted.value_counts(dropna=False))
print("-"*100)
```

```python
def get_rates(column, data):
    print('{} Readmission Rates'.format(column))
    print('-'*25)
    readmission_rates = []
    counts = []
    differences = []
    unique_values = data[column].unique()
    for unique_value in unique_values:
        if type(unique_value) == float:
            readmission_rate = data[(data[column].isna()) & (data.readmitted == 1)].shape[0] \
                               / data[data[column].isna()].shape[0]
        else:
            readmission_rate = data[(data[column] == unique_value) & (data.readmitted == 1)].shape[0] \
                               / data[data[column] == unique_value].shape[0]
        readmission_rates.append(readmission_rate)
        counts.append(data[data[column] == unique_value].shape[0])
        differences.append(readmission_rate - (data[data.readmitted == 1].shape[0]/data.shape[0]))
    readmission_rate_df = pd.DataFrame({'Value': unique_values, 'Rate':readmission_rates,\
                                        'Counts' : counts, 'Difference': differences})
    readmission_rate_df = readmission_rate_df.sort_values(by=['Rate'], ascending=False)
    print(readmission_rate_df)
    print('')
    print('')
```

```python
variables = list(data.columns)
variables.remove('encounter_id')
variables.remove('patient_nbr')
variables.remove('weight')
variables.remove('diag_1')
variables.remove('diag_2')
variables.remove('diag_3')
variables.remove('readmitted')
```

```python
print('Total Readmission Rate: ', (data[data.readmitted == 1].shape[0]/data.shape[0]))

for variable in variables:
    get_rates(variable, data)
```
